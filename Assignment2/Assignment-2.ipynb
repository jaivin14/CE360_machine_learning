```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

## Spacing using padding and tight_layout

```python
# plot 1:
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,2,1) #the figure has 2 row, 2 column, and this is first plot
plt.title("PLOT 1 : ")
plt.plot(x,y)

# plot 2:
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,2,2) # the figure has 2 row and 2 columns and this is second plot
plt.title("PLOT 2 : ")
plt.plot(x,y)

# plot 3:
x = np.array([0,1,2,3])
y = np.array([50,40,30,20])

plt.subplot(2,2,3) # the figure has 2 row and 2 columns and this is third plot
plt.title("PLOT 3 : ")
plt.plot(x,y)

# plot 4:
x = np.array([0,1,2,3])
y = np.array([20,40,60,50])

plt.subplot(2,2,4) # the figure has 2 row and 2 columns and this is fourth plot
plt.title("PLOT 4 : ")
plt.plot(x,y)
plt.suptitle("SUPER TITLE")
plt.tight_layout(pad=1.0)
plt.show()
```

## Overlaped bar chart

```python
a = range(1,10)
b = range(4,13)
Xindex = np.arange(len(a))

fig = plt.figure()
barChart = fig.add_subplot(1,1,1)
barChart.bar(x=Xindex, height=a, width=0.4,align='center')
barChart.bar(x=Xindex, height=b, width=0.2,align='center')

plt.tight_layout()
plt.show()
```

## stacked bar chart

```python
import matplotlib.pyplot as plt
x = ['Cat_1', 'Cat_2', 'Cat_3', 'Cat_4']
y1 = [16, 30, 38, 24]
y2 = [19, 35, 14, 35]

plt.bar(x, y1, color='g')
plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```
